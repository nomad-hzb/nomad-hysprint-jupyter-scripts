Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from ipywidgets import interact, FloatRangeSlider, Button, VBox, HBox, Output
```

Settings

```python
# === SETTINGS ===
input_folder = "./xy_files/"
output_folder = "./aligned_xy_files/"
align_to = "first"  # or 'mean'
```

File import + Reading Files + Plotting raw data 

```python
# Create output folder if needed
os.makedirs(output_folder, exist_ok=True)

# Load all .xy files
file_paths = sorted(glob.glob(os.path.join(input_folder, "*.xy")))
spectra_raw = []

for path in file_paths:
    data = np.loadtxt(path)
    x, y = data[:, 0], data[:, 1]
    spectra_raw.append((x, y))

# Get global x-range for slider
x_global = np.concatenate([x for x, _ in spectra_raw])
x_min_slider = np.min(x_global)
x_max_slider = np.max(x_global)

# === STEP 1: Plot raw spectra ===
print("Showing raw spectra:")
plt.figure(figsize=(8, 5))
for x, y in spectra_raw:
    plt.plot(x, y)
plt.title("Raw Spectra (Not Normalized)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

Define Slider in step 2

```python
# === STEP 2: INTERACTIVE WIDGET TO SET PEAK INTERVAL ===
out = Output()

interval_slider = FloatRangeSlider(
    value=[x_min_slider + 0.1*(x_max_slider - x_min_slider), x_max_slider - 0.1*(x_max_slider - x_min_slider)],
    min=x_min_slider,
    max=x_max_slider,
    step=0.1,
    description='Peak range:',
    continuous_update=False,
    layout={"width": "80%"}
)

run_button = Button(description="Align and Normalize", button_style='success')

def on_button_clicked(b):
    with out:
        out.clear_output()

        # Get selected interval
        x_min, x_max = interval_slider.value

        # Normalize, find peaks, and align
        peak_positions = []
        spectra_norm = []

        for x, y in spectra_raw:
            y_norm = y / np.max(y)
            spectra_norm.append((x, y_norm))

            mask = (x >= x_min) & (x <= x_max)
            if not np.any(mask):
                raise ValueError(f"No data in interval [{x_min}, {x_max}]")
            peak_idx = np.argmax(y_norm[mask])
            peak_x = x[mask][peak_idx]
            peak_positions.append(peak_x)

        if align_to == "first":
            ref_peak = peak_positions[0]
        elif align_to == "mean":
            ref_peak = np.mean(peak_positions)
        else:
            raise ValueError("Invalid align_to value.")

        # Apply shift
        spectra_aligned = []
        for (x, y), peak_x in zip(spectra_norm, peak_positions):
            x_shifted = x + (ref_peak - peak_x)
            spectra_aligned.append((x_shifted, y))

        # Plot original normalized
        fig, axs = plt.subplots(1, 2, figsize=(14, 5))

        axs[0].set_title("Original Spectra (Normalized)")
        for x, y in spectra_norm:
            axs[0].plot(x, y)
        axs[0].set_xlabel("x")
        axs[0].set_ylabel("Normalized y")
        axs[0].grid(True)

        # Plot aligned normalized
        axs[1].set_title("Aligned Spectra (Normalized)")
        for x, y in spectra_aligned:
            axs[1].plot(x, y)
        axs[1].set_xlabel("x (shifted)")
        axs[1].set_ylabel("Normalized y")
        axs[1].grid(True)

        plt.tight_layout()
        plt.show()

        # Save aligned spectra
        for (x, y), original_path in zip(spectra_aligned, file_paths):
            filename = os.path.basename(original_path)
            output_path = os.path.join(output_folder, filename)
            np.savetxt(output_path, np.column_stack([x, y]), fmt="%.6f")

        print(f"\nDone! Aligned files saved to: {output_folder}")

# Attach handler
run_button.on_click(on_button_clicked)

# Display widgets
display(VBox([interval_slider, run_button, out]))
```
